# Actividad 3: Pivot Tables

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head(2)
```

```python
pivot = pd.pivot_table(df, values = ['peso'], index = ['desc_ent'], aggfunc ={'peso':np.mean})
pivot
```

```python
pivot2 = pd.pivot_table(df, values = ['peso'], index = ['desc_ent'], columns = 'sexo', aggfunc ={'peso':np.mean})
pivot2
```

```python
df2 = pd.read_csv('datasets/adult.csv')
df2.head(2)
```

```python
pivot3 = pd.pivot_table(df2, index = ['education', 'marital-status'], columns = ['sex'], values = ['age', 'hours-per-week'])
pivot3.head(60)
```
