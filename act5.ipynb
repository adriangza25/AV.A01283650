#  Actividad 5: Visualización con Histogramas & Boxplots

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head(2)
```

```python
sns.boxplot( x=df["cintura"],) #showfliers = False)
```

```python
pivot = pd.pivot_table(df, index = ['desc_ent'], columns = ['sexo'], values = ['edad'], aggfunc = {'edad':['median']})
pivot
```

```python
pivot.boxplot(column = [('edad','median',2)])
```

```python
pivot = pd.pivot_table(df, index = ['desc_ent'], columns = ['sexo'], values = ['cintura'], aggfunc = {'cintura':['median']})
pivot
```

```python
pivot[('cintura', 'median', 2)].plot(kind = "line")
```

```python
pivot_sistol = pd.pivot_table(df, index = ['desc_ent'], columns = ['sexo'], values = ['sistol'], aggfunc = {'sistol':['sum']})
pivot_sistol
```

```python
s = pivot_sistol.sort_values(('sistol', 'sum', 1), ascending = False)
s10 = s.head(10)
```

```python
s10.plot(kind = 'barh')
```
