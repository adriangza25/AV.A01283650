```python
import statsmodels.api as sm
```

```python
df = pd.read_csv('datasets/nhanes_2015_2016.csv')
df.head()
```

```python
vars = ['RIDAGEYR', 'BPXSY1']
df = df[vars].dropna()
df
```

```python
df[vars].corr()
```

```python
model = sm.OLS.from_formula('BPXSY1 ~ RIDAGEYR',
                           data = df) #BPXSY1 = b0 + b1*RIDAGEYR
result = model.fit()
result.summary()
```

```python
x = result.fittedvalues
y = result.resid
```

```python
fig = sns.scatterplot(x = x,
                     y = y,
                     )
fig.set_xlabel('Fitted Values')
fig.set_ylabel('Residuals')
```
