```python
# np.arange (start,end,step)
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y)
```

```python
ysin = np.sin(x)
ycos = np.cos(x)

plt.plot(x, ysin)
plt.plot(x, ycos)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sine and Cosine funcitons')

plt.show()
```

```python
plt.subplot(2, 1, 1) #2 rows, 1 column, position 1
plt.plot(x, ysin)
plt.title('Sine')

plt.subplot(2, 1, 2) #2 rows, 1 column, position 2
plt.plot(x, ycos)
plt.title('Cosine')

plt.show()
```

```python
import seaborn as sns
```

```python
path = 'datasets/Cartwheeldata.csv'
df = pd.read_csv(path)
df.head(2)
```

```python
sns.lmplot(x = 'Wingspan',
           y = 'Height',
           #fit_reg = False,
           #hue = 'Gender',
           data = df)
plt.title('Scatter plot: Wingspan vs Height')
plt.show()
```

```python
sns.swarmplot(x = 'Gender',
              y = 'Wingspan',
              data = df,)
plt.title('Swarmplot: Gender vs Wingspan')
plt.show()
```

```python
sns.boxplot(data = df.loc[df['Gender'] == 'F', 
                          ['Age', 'Height', 'Wingspan', 'CWDistance',]])
plt.title('Female boxplot')
```
