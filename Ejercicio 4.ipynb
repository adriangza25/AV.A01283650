```python
import statsmodels.api as sm
```

```python
df = pd.read_csv('datasets/nhanes_2015_2016.csv')
df.head()
```

```python
vars = ['RIDAGEYR', 'BPXSY1']
df = df[vars].dropna()
df
```

```python
model = sm.OLS.from_formula('BPXSY1 ~ RIDAGEYR',
                           data = df) #BPXSY1 = b0 + b1*RIDAGEYR
result = model.fit()
result.summary()
```

```python
ax = sns.regplot(x = 'BPXSY1',
                 y = 'RIDAGEYR',
                data = df,
                x_estimator = np.mean,
                ci = 95,
                 
                )
ax
```
