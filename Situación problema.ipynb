```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_excel('datasets/Happiness Index.xlsx',
                  engine = 'openpyxl')
df.head(2)
```

```python
df.set_index('Overall rank', inplace = True)
df.head(2)
```

```python
sns.distplot(df.Score)
plt.title('"Normality" of worldwide happiness level')
plt.xlabel('Happiness score')
```

```python
top = df.sort_values(by = 'Score',
                       ascending = False,
                       axis = 0
                       )
top = top.head(5)
top
```

```python
df.columns
```

```python
sns.lmplot(x = 'Score',
           y = 'Healthy life expectancy',
           #fit_reg = False,
           data = df)
plt.title('Scatter plot: Healthy life expectancy vs Score')
plt.show()
```

```python
sns.lmplot(x = 'Score',
           y = 'Perceptions of corruption',
           #fit_reg = False,
           data = df)
plt.title('Scatter plot: Perceptions of corruption vs Score')
plt.show()
```

```python
sns.lmplot(x = 'Score',
           y = 'Generosity',
           #fit_reg = False,
           data = df)
plt.title('Scatter plot: Generosity vs Score')
plt.show()
```

```python
sns.lmplot(x = 'Score',
           y = 'Freedom to make life choices',
           #fit_reg = False,
           data = df)
plt.title('Scatter plot: Freedom to make life choices vs Score')
plt.show()
```

```python
sns.lmplot(x = 'Score',
           y = 'Social support',
           #fit_reg = False,
           data = df)
plt.title('Scatter plot: Social support vs Score')
plt.show()
```

```python
sns.swarmplot(x = 'Continent',
              y = 'Score',
              data = df)
plt.title('Swarmplot: Continent vs Score')
plt.show()
```
