```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_excel('datasets/Happiness Index.xlsx',
                  engine = 'openpyxl')
df.head(2)
```

```python
df.set_index('Overall rank', inplace = True)
df.head(2)
```

```python
histogram = df['Score'].plot(kind = 'hist')
plt.title('"Normality" of worldwide happiness level')
plt.xlabel('Happiness score')
```

```python
sns.lmplot(x = 'Score',
           y = 'GDP per capita',
           #fit_reg = False,
           data = df)
plt.title('Scatter plot: GDP per capita vs Score')
plt.show()
```
